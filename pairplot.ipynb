### **PairPlot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.pairplot(df, hue="sex")
plt.show()
```

**Using var**

```python
sns.pairplot(df, vars=["total_bill", "tip"], hue='sex')
plt.show()
```

**Kind**

```python
sns.pairplot(df, hue="sex", kind="kde") #kind - scatter, kde, hist, reg
plt.show()
```

**diag_kind**

```python
sns.pairplot(df, hue="sex", diag_kind="hist")
plt.show()
```

**Marker**

```python
sns.pairplot(df, hue="sex",markers=["*", ">"])
plt.show()
```
