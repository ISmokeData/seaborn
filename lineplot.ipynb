### **LinePlot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7,8]
var_1 = [2,3, 4, 5, 1, 5 , 6, 7]
```

```python
plt.plot(var, var_1)
plt.show()
```

```python
x = pd.DataFrame({"var":var, "var_1":var_1})
sns.lineplot(x = "var", y= "var_1",data=x )
plt.show()
```

```python
df = sns.load_dataset("penguins")
```

```python
df.head()
```

```python
sns.lineplot(x = df['bill_depth_mm'], y = df['flipper_length_mm'], data=df )
plt.show()
```

```python
sns.lineplot(x = df['bill_depth_mm'], y = df['flipper_length_mm'], data=df, hue= df['sex'] )
plt.show()
```

**style**

```python
sns.lineplot(x = df['bill_depth_mm'], y = df['flipper_length_mm'], data=df , style= df['sex'])
plt.show()
```

**Patlette**

```python
sns.lineplot(x = df['bill_depth_mm'], y = df['flipper_length_mm'], data=df,hue=df['sex'],style="sex", palette="Accent" )
plt.show()
```

**Marker**

```python
sns.lineplot(x = df['bill_depth_mm'], y = df['flipper_length_mm'], data=df,hue=df['sex'],style="sex", palette="Accent", markers=["o", ">"] )
plt.show()
```

**Dash** remove dotted line

```python
sns.lineplot(x = df['bill_depth_mm'], y = df['flipper_length_mm'], 
             data=df,hue=df['sex'],style="sex", palette="Accent", dashes=False )
plt.show()
```

**Remove Instruction** -- use legend

```python
sns.lineplot(x = df['bill_depth_mm'], y = df['flipper_length_mm'],
             data=df,hue=df['sex'],style="sex", palette="Accent", legend=False )
plt.show()
```

```python
sns.lineplot(x = df['bill_depth_mm'], y = df['flipper_length_mm'], 
             data=df,hue=df['sex'],style="sex", palette="Accent" )
plt.grid()
plt.show()
```
