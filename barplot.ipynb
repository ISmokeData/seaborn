### **BarPlot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
```

```python
df.head()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex")
plt.show()
```

**Order change**

```python
order = ["Dream", "Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", order=order)
plt.show()
```

**Change hue order**

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", hue_order=["Female", "Male"])
plt.show()
```

**Black dot on bar**

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", ci=50)
plt.show()
```

**Horizontal** - both parameter must be numerical

```python
sns.barplot(y="bill_depth_mm", x="bill_length_mm", data=df, orient="h")
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", palette="OrRd")
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", palette="OrRd",saturation=0.5)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", palette="OrRd", errcolor="g")
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", palette="OrRd", errcolor="g", errwidth=5)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", palette="OrRd", errcolor="g", capsize=0.3)
plt.show()
```

**SuperImpose graph**

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", palette="OrRd", errcolor="g",dodge=False)
plt.show()
```

```python
sns.set(style = "darkgrid")
sns.barplot(x="island", y="bill_length_mm", data=df, hue= "sex", errcolor="g")
plt.show()
```
