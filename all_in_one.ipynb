### Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatter -> axis level function
sns.color_palette("pastel")
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# scatter -> figure level finction
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style='darkgrid', palette='Set2', font_scale=1.2)
sns.scatterplot(x='total_bill', y='tip', data=tips,
                hue='sex', style='time', size='size')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips,
            kind='scatter', hue='sex', style='time')
```

### pair grid

```python
g = sns.PairGrid(data=iris, hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species', vars=['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
