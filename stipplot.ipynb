### **Strip Plot**
- A strip plot is very simple to understand. It is baiscally a scatter plot that differentiates different categories. Strip plots are considered a good alternative to a box plot or a violin plot.

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue= "sex",edgecolor="black")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue= "sex",edgecolor="black", jitter=2)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue= "sex",edgecolor="black", size=8)
plt.show()
```

```python
mark={"Male":"o","Female":">"}
sns.stripplot(x="day", y="total_bill", data=df, hue= "sex",size=7,marker=">")
plt.show()
```

```python
sns.stripplot(y=df['total_bill'])
plt.show()
```
