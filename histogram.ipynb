### **Histogram**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
```

```python
df.head()
```

```python
sns.displot(df['bill_depth_mm'])
plt.show()
```

**Bins** - organised bars

```python
sns.displot(df['flipper_length_mm'], bins=[170, 180, 190, 200, 210, 220, 230, 240])
plt.show()
```

**kernal density**

```python
sns.displot(df['flipper_length_mm'], kde = True)
plt.show()
```

**Rug**

```python
sns.displot(df['flipper_length_mm'], kde =True, rug=True)
plt.show()
```

```python
sns.displot(df['flipper_length_mm'], kde =True, rug=True, color= "r")
plt.show()
```
