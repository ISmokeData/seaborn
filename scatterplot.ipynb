### **ScatterPlot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
```

```python
df.head()
```

```python
sns.scatterplot(x = df['bill_length_mm'], y = df['bill_depth_mm'], data=df, hue="sex")
plt.show()
```

**style**

```python
sns.scatterplot(x = df['bill_length_mm'], y = df['bill_depth_mm'], data=df, hue="sex", style="sex")
plt.show()
```

**size**

```python
sns.scatterplot(x = df['bill_length_mm'], y = df['bill_depth_mm'], data=df, 
                hue="sex", style="sex", size= "sex")
plt.show()
```

```python
sns.scatterplot(x = df['bill_length_mm'], y = df['bill_depth_mm'], data=df, 
                hue="sex", style="sex", size= "sex", palette="OrRd")
plt.show()
```
